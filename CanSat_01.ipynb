```
import numpy as np
from skimage.io import imread
import matplotlib.pyplot as plt
from skimage.measure import label
import cv2
from google.colab.patches import cv2_imshow
```

```
image = imread('./target.PNG')
image = np.mean(image, axis=2)

plt.figure(figsize=(15, 20))
plt.subplot(121), plt.imshow(image, cmap='gray')
plt.subplot(122), plt.imshow(image[70:170, 600:800], cmap='gray')
```

```
plt.figure(figsize=(15, 20))
plt.subplot(121), plt.imshow(image > 220, cmap='gray')
plt.subplot(122), plt.imshow(image[70:170, 600:800] > 220, cmap='gray')
```

```
image = image[70:170, 600:800]
thresh = cv2.threshold(image, 220, 255, cv2.THRESH_BINARY)[1]
contours, hierarchy = cv2.findContours(thresh.astype(np.uint8), 1, 2)
cnt = contours[0]

rects = [cv2.boundingRect(cnt) for cnt in contours]
rects = sorted(rects,key=lambda  x:x[1],reverse=True)

print(rects)
```
